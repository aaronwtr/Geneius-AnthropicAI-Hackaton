```python
import numpy as np
import pandas as pd
import anndata as ad
import shap
from copy import copy
from sklearn.model_selection import StratifiedKFold
from xgboost import XGBClassifier
import matplotlib as mpl
from matplotlib import colors as plt_colors
import matplotlib.pyplot as plt
%matplotlib inline

cell_type_map = {'Mono_CD14':0, 'Mono_CD16':1, 'Macro_SPP1':2, 'cDC2_CD1C':3,
                 'Macro_C1QC':4, 'Macro_LYVE1':5, 'Macro_NLRP3':6, 'cDC3_LAMP3':7,
                 'Macro_INHBA':8, 'cDC1_CLEC9A':9, 'pDC_LILRA4':10}
inverted_cell_type_map = {v: k for k, v in cell_type_map.items()}

cancer_type_map = {'ESCA':0, 'MYE':1, 'OV-FTC':2}
inverted_cancer_type_map = {v: k for k, v in cancer_type_map.items()}
```

# Load Data:

```python
mye_data = ad.read_h5ad('../data/mye/mye_query_adata.h5ad')
print(mye_data.shape)
mye_data
```

# Cross-Val Splitting:

```python
# Define feature and target data:
X_data = mye_data.X

# Extract the cell type and convert to integer values:
label = 'cancer'
if label == 'cell':
    y_vals = mye_data.obs.cell_type.values
    y_target = np.array([cell_type_map[cell_type] for cell_type in y_vals])
elif label == 'cancer':
    y_vals = mye_data.obs.cancer_type.values
    y_target = np.array([cancer_type_map[cancer_type] for cancer_type in y_vals])
```

```python
# Define the number of folds and test size:
n_splits = 5
test_size = 0.2

# Initialize the StratifiedKFold:
skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=42)

# Initialize lists to store train and test indices
train_indices = []
test_indices = []

# Perform the split:
for train_index, test_index in skf.split(X_data, y_target):
    train_indices.append(train_index)
    test_indices.append(test_index)
    
cv_data = []

# Split data according to fold indices:
for fold, (train_idx, test_idx) in enumerate(zip(train_indices, test_indices)):
    X_train, y_train = X_data[train_idx], y_target[train_idx]
    X_test, y_test = X_data[test_idx], y_target[test_idx]
    cv_data.append({'X_train': X_train, 'y_train': y_train, 
                   'X_test': X_test, 'y_test': y_test})

    print(f"Fold {fold + 1}:")
    print(f"Train Set - X shape: {X_train.shape}, y shape: {y_train.shape}")
    print(f"Test Set - X shape: {X_test.shape}, y shape: {y_test.shape}")
    print()
```

# Shapley Feature Importance:

```python
def shapely_cv(cv_data, model, feat_names):
    
    cv_cols = []
    cv_shap_vals = []
    
    # Return cross-validated data splits:
    for cv in range(len(cv_data)):
        print("Beginning CV %s..." % (cv+1))
        X_train, X_val, y_train, y_val = cv_data[cv]['X_train'], cv_data[cv]['X_train'], \
                                            cv_data[cv]['y_train'], cv_data[cv]['y_test']
        
        # Copy model:
        cv_mod = copy(model)
        # Fit the model on the training data:
        cv_mod.fit(X_train, y_train)
        
        # Define tree explainer:
        mod_exp = shap.TreeExplainer(cv_mod)
        # Generate Shapely values using test set:
        shap_vals = mod_exp.shap_values(X_val)
        cv_shap_vals.append(shap_vals)
        # Dataframe of shapley values per feature:
        shap_df, cv_col = global_shap_importance(shap_vals, feat_names, cv)
        cv_cols.append(cv_col)
        
        if cv == 0:
            res_df = shap_df
        else:
            res_df = pd.merge(res_df, shap_df, on='features')
            
    # Calculate mean and std across cv folds:
    res_df['cv_std'] = res_df[cv_cols].std(axis=1)
    res_df['cv_mean'] = res_df[cv_cols].mean(axis=1)
     
    # Sort by mean shapley values:
    res_df.sort_values(by=['cv_mean'], ascending=False, inplace=True)
    
    return res_df, cv_shap_vals
```

```python
def global_shap_importance(shap_values, feat_names, cv=''):
    """ Return a dataframe containing the features sorted by Shap importance
    Parameters
    ----------
    shap_values : The shapely values produced by the model
    feat_names : feature names as a list
    Returns
    -------
    pd.Dataframe
        A dataframe containing the features sorted by mean Shapley value (importance to the model)
    """
    vals_abs = np.abs(shap_values)
    val_mean = np.mean(vals_abs, axis=1) # average over test datapoints
    val_final = np.sum(val_mean, axis=0) # sum over classes
    sv_col = 'sv_%s' % cv
    feature_importance = pd.DataFrame(
                        list(zip(feat_names, val_final)), columns=['features', sv_col])
    
    return feature_importance, sv_col
```

# Shapley XGBoost:

```python
# Define model:
xgb_model = XGBClassifier()
```

```python
# Get feature (Gene) names:
mye_genes = list(np.array(mye_data.var.index))
mye_genes[0:5]
```

```python
# Get CV results using validation sets:
cv_results, cv_sv = shapely_cv(cv_data, xgb_model, mye_genes)
```

```python
cv_results = cv_results.reset_index(drop=True)
cv_results['feature_n'] = cv_results.index + 1
cv_results.head()
```

```python
cv_results.to_csv('../results/cv_res/cancer_type_cv_res.csv', index=False)
```

# Summary Plot of Results:

```python
def shap_summary_plot(shap_vals, feat_cols, class_dict, title, rgb_cm):
    # Initialize Shapely plot:
    # Custom colormap based on class indices in shap values:
    class_inds = np.argsort([-np.abs(shap_vals[i]).mean() for i in range(len(shap_vals))])
    cmap = plt_colors.ListedColormap(np.array(rgb_cm)[class_inds])

    shap.summary_plot(shap_vals, feat_cols, show = False, color=cmap)

    # Get the Shapely figure and axes objects.
    fig, ax = plt.gcf().set_size_inches(12,6), plt.gca()

    # Adjust legened:
    handles, labels = plt.gca().get_legend_handles_labels()
    # Get class indices in order from summary_plot:
    class_idx = [val for sublist in [[int(s) for s in x.split() if s.isdigit()] for x in labels] for val in sublist]
    # Zip together handles and class indices and sort by index:
    zip_hl = sorted(zip(handles, class_idx), key=lambda x: x[1])
    sorted_hdls = [h[0] for h in zip_hl]
    # Return class names as labels according to index:
    sorted_lbls = [class_dict[x] for x in [l[1] for l in zip_hl]]

    # Modifying main plot parameters
    ax.tick_params(labelsize=9)
    ax.set_xlabel("Mean SHAP value (Impact on model output)", fontsize=10)
    ax.set_title(title, fontsize=12, fontweight='bold')
    plt.legend(fontsize=10, handles= sorted_hdls, labels = sorted_lbls)

    plt.show()
```
